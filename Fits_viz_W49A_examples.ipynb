## In this notebook, we give an overview of the  fits visualization template module using published IRS SL data of the star forming region W49A

```python
#First include path to fits_viz_template in path environment
import sys
sys.path.insert(0,'pyscripts/')
```

```python
# Import function to output quick visualizations with fits input
from fits_viz_template import fits_viz as viz
```

```python
# Set some default parameters
fitspath = 'IRS/measure_corr/Fluxes/' 
upper_p = 99
lower_p = 5
scaling = "linear"
```

```python
#import relevant contours
contourname_PAH77 = 'IRS/measure_corr/Fluxes/w49_map77_sl.fits'
contour_levels_PAH77 =[3e-5, 4.8e-5, 5.7e-5, 9.5e-5, 1.4e-4, 1.6e-4]
contourname_PAH112 = 'IRS/measure_corr/Fluxes/w49_map112_sl.fits'
contour_levels_PAH112 =[1.2e-5, 1.9e-5, 2.6e-5, 4.3e-5, 6.6e-5]
tickspacing = [2.7777777e-4*2, 15*2]
```

#### 1. Basic function call

```python
# plot IRS SL PAH 6.2 map with minimal parameters defined 
filename ='IRS/measure_corr/Fluxes/w49_map62_sl.fits'
viz(filename= filename)
```

#### Using only the filename as input, we can output basic header info and plot a map with an associated colorbar. 

#### 2. We can also adjust input parameters slightly to output histrogram showing the distribution of values in our data.

```python
# plot IRS SL PAH 6.2 histrogram and turn off image plotting
filename ='IRS/measure_corr/Fluxes/w49_map62_sl.fits'
viz(filename= filename, hist_display=True, im_display = False)
```

```python
# Output IRS SL PAH 6.2 map to file (currently only .png files) 
filename ='IRS/measure_corr/Fluxes/w49_map62_sl.fits'
viz(filename= filename,plotpath=fitspath,save=True)
```

#### 3. Adjust the range, scaling, and colormap  shown in map

```python
# Plot IRS SL PAH 6.2 map from the 10th to 90th percentile 
# using a square root scaling and alternate color map
filename ='IRS/measure_corr/Fluxes/w49_map62_sl.fits'
viz(filename= filename,upper_p=90, lower_p=10,
    scaling= "sqrt", cmap = "viridis")
```

#### 4. Output simple statistics with units given in header

```python
# output IRS SL PAH 6.2 statistics and get units from fits header
filename ='IRS/measure_corr/Fluxes/w49_map62_sl.fits'
viz(filename= filename,show_stats= True, extract_units=True, im_display = False)
```

#### 5. Use wcs information to overplot contours from other fits files

```python
# Plot IRS SL PAH 6.2 map with contours from PAH 7.7 and PAH 11.2
viz(filename= filename, upper_p=upper_p, lower_p=lower_p, scaling= scaling,
    contourname1=contourname_PAH112, contour_levels1=contour_levels_PAH112,
    contourname2=contourname_PAH77, contour_levels2=contour_levels_PAH77)
```

#### 6. Create sub images and output them to fits files

```python
# Plot a subimage of IRS SL PAH 6.2 and save it as a new fits file
viz(filename= filename, plotpath =fitspath,
    a1 =30, a2 = 73, b1 =0, b2 = 70, save_fits = True)
```

#### 7. Output map with wcs as axes and coordinate grid overplotted (currently only RA and DEC supported)

```python
# plot IRS SL PAH 6.2 
viz(filename= filename, plotpath=fitspath,
    upper_p=upper_p, lower_p=lower_p, scaling= scaling,
    wcs_axes = True, rot_wcs =True, grid = True,
    tickspacing = tickspacing,
    contourname1=contourname_PAH112, contour_levels1=contour_levels_PAH112,
    contourname2=contourname_PAH77, contour_levels2=contour_levels_PAH77)
```

#### Note the rotation of the RA and DEC axes here  due to the orientation of the image grid, in most cases these would be switched. Here the rot_wcs keyword takes care of this rotation.
